### Light Conditions vs Growth Time Analysis

```python
sb.countplot(x='Light_Cat',data=dt)
plt.title('Distribution of Light Conditions')
plt.show()
```

```python
sb.boxplot(x='Light_Cat', y='Time', data=dt)
plt.title('Growth Time vs Light Conditions')
plt.show()
```

Overall, tree under the Medium light condition has longer growth time while that under low and high ligh condition only occur much higher growth time occasionally.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
lt_mdl=ols('Time~C(Light_Cat)',data=dt).fit()
anova_table = sm.stats.anova_lm(lt_mdl, typ=2)
anova_table
```

**Null Hypothesis**: There are no differences in the means of the 'Time' variable across the different levels of 'Light_Cat'.\
**Alternative Hypothesis**: There are significance in the means of the 'Time' variable across the different levels of 'Light_Cat'.\
**Result**: The P-value much fewer than 0.01, suggesting that there's a statistically significant effect of light conditions on growth time, with at least one group's mean significantly different from the others. Given the low p-value, we reject the null hypothesis of equal means across light conditions.

```python
plt.figure(figsize=(12, 8))
sb.barplot(x=Soil_ct.index,y=Soil_ct.values)
plt.xlabel('Type of Soil')
plt.ylabel('Count')
plt.title('Distribution of Soil')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sb.boxplot(x='Soil', y='Time', data=dt)
plt.title('Growth Time vs Soil type')
plt.show()
```

From the graphic, we can tell that the type of soil might be impact on the time of tree growing. In order to check the validity of this assumption, the ANOVA test can be helpful.

```python
st_mdl=ols('Time~C(Soil)',data=dt).fit()
st_anova_table = sm.stats.anova_lm(st_mdl, typ=2)
st_anova_table
```

**Null Hypothesis**: There are no differences in the means of the 'Time' variable across the different types of 'Soil'.\
**Alternative Hypothesis**: There are significance in the means of the 'Time' variable across the different types of 'Soil'.\
**Result**: Since the P-value is much greater than 0.05, which means we have not enough evidence to reject the null hypothesis that there are no differences between types of soil, we now can say the soil variable might have no impact on the time of tree growing. 

### Time Series Analysis of Census Data

```python
dt['PlantDate']=pd.to_datetime(dt['PlantDate'])
```

```python
date_cen_group=dt.groupby('PlantDate')['Census'].mean()
date_cen_group
```

```python
time_dt=dt[['PlantDate','Census']]
time_dt.set_index('PlantDate',inplace=True)
```

```python
mean_census = dt['Census'].mean()
std_dev_census = dt['Census'].std()
```

```python
plt.figure(figsize=(10, 6))  # You can adjust the size as needed
sb.lineplot(data=time_dt, x=time_dt.index, y='Census', label='Census')

plt.axhline(mean_census, color='green', linestyle='--', label='Mean')

plt.axhline(mean_census + std_dev_census, color='red', linestyle='--', label='Mean + 1 SD')
plt.axhline(mean_census - std_dev_census, color='blue', linestyle='--', label='Mean - 1 SD')
plt.xticks(rotation=45)
plt.xlabel('PlantDate')
plt.ylabel('Census')
plt.title('Census over Time with Mean and Standard Deviation')
plt.legend()

# Display the plot
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
adft = adfuller(time_dt,autolag="AIC")
```

```python
output_dt = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]  , "Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"]})
print(output_dt)
```

**Null Hypothesis**: The time series has a unit root, indicating it is non-stationary.\
**Alternative Hypothesis**: The time series does not have a unit root, indicating it is stationary.\
**Result**: Since the P-value is much lower than 0.05, indicating that the time series here doesn't have unit root and it's stationary.

### Correlation Analysis between Key Variables

```python
dt_cor=dt[['Light_ISF','AMF','EMF','Phenolics','Lignin','NSC','Census','Time','Event']]
cormat=dt_cor.corr()
sb.heatmap(cormat, annot=True)
```

```python
cormat
```

**Result**:\
From the observation of the heat map, we can find that  'Phenolics', 'Lignin', 'EMF', and 'NSC' might be important predictors due to their stronger correlations with the 'Event' variable. However, it's weird that 'EMF' has value of 1 correlation with the Event, which is the variable that we interest in predicting, so we should not put it into the prediction model. Also, the Phenolics variable have strong correlation with both Lignin and NSC variable, might be the multicollinearity issue. So we should take a test of them.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 
dt_colintest=dt_cor.drop(columns='Event')
```

```python
vif_data = pd.DataFrame() 
vif_data["feature"] = dt_colintest.columns 
vif_data["VIF"] = [variance_inflation_factor(dt_colintest.values, i) 
                          for i in range(len(dt_colintest.columns))] 
  
print(vif_data)
```

**Result**:\
The variables Light_ISF, Lignin, and NSC showing high value of VIF, indicating that these three variables should be considered before put into the model since they might have multicollinearity issue.

### Cluster Analysis for Pattern Identification

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
features=dt[['Light_ISF', 'AMF', 'Phenolics', 'Lignin', 'NSC']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

**Result**:\
We can see the since the value 3 the line starts leveling off, so I would set the optimal_number_of_clusters to be 3.

```python
optimal_number_of_clusters=3
kmeans = KMeans(n_clusters=optimal_number_of_clusters, random_state=0)
cluster_labels = kmeans.fit_predict(scaled_features)
```

```python
dt['Cluster'] = cluster_labels

print(dt.groupby('Cluster').mean())
```

**Means of Features by Cluster**:
  - **Cluster 0**: This group has the highest mean values for 'Census' and 'Event', suggesting that trees in this cluster have a higher count and a higher occurrence of the event you're interested in.
  - **Cluster 1**: This group has lower mean values for 'Census' and 'Event' compared to Cluster 0, indicating fewer counts and events.
  - **Cluster 2**: This group has the lowest mean 'Census' but the highest mean 'AMF', which might suggest a particular type of soil or environmental condition where these trees are located.
    
**Differences in Other Features**:
  - **'Core'** is approximately the same across all clusters, which indicates it might not be a differentiating factor.
  - **'AMF'** is notably higher in Cluster 2, which might have some ecological or environmental implications.
  - **'EMF', 'Phenolics', 'Lignin', and 'NSC'** also vary across clusters. For instance, **'Phenolics' and 'Lignin'** are highest in Cluster 0, which could suggest differences in tree species, age, or health.
  
**Interpretation for Prediction**:
  - If 'Event' is the outcome you want to predict, Cluster 0 might be of particular interest because it has the highest mean 'Event' value.
  - The differences in the mean values of features across clusters might be indicative of which features are important for predicting 'Event'. For example, if Cluster 0 has higher 'Event' values and also has higher means for certain features, those features might be good predictors for the event occurring.

## Tryout1- Classification / Decision Tree

```python
print(dt['Phenolics'].min(),
dt['Phenolics'].max())
```

```python
dt = dt[dt['Event'].isin([0, 1])]
dt['Pheno_cat']=0
dt.loc[dt['Phenolics']>1.15,'Pheno_cat']=1
dt.loc[dt['Phenolics']>3.63,'Pheno_cat']=2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y=dt[['Event']]
X_encoded = pd.get_dummies(dt[['Species', 'Light_Cat', 'Soil', 'Sterile','Pheno_cat']], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
cl_mdl= DecisionTreeClassifier(random_state=0)
cl_mdl.fit(X_train, y_train)
cl_pred = cl_mdl.predict(X_test)
```

```python
print("Accuracy of Decision Tree Classifier: ",
      accuracy_score(y_test, cl_pred))
print("Precision of Decision Tree Classifier: ",
      precision_score(y_test, cl_pred, average='weighted'))
print("Recall of Decision Tree Classifier: ",
      recall_score(y_test, cl_pred, average='weighted'))
print("F1-Score of Decision Tree Classifier: ",
      f1_score(y_test, cl_pred, average='weighted'))
```

```python
print(classification_report(y_test, cl_pred))
```

```python
clf = svm.SVC()
clf.fit(X_train, y_train)
clf_pred = clf.predict(X_test)
print("Accuracy of SVM Classifier: ",
      accuracy_score(y_test, clf_pred))
print("Precision of SVM Classifier: ",
      precision_score(y_test, clf_pred, average='weighted'))
print("Recall of SVM Classifier: ",
      recall_score(y_test, clf_pred, average='weighted'))
print("F1-Score of SVM Classifier: ",
      f1_score(y_test, clf_pred, average='weighted'))
```

```python
print(classification_report(y_test, clf_pred))
```
