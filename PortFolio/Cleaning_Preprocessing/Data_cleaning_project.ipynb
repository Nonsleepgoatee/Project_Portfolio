```python
import re
def height_converter(height):
    match_ftich=re.match(r'(\d+)\'(\d+)\"', height)
    if match_ftich:
        feet=int(match_ftich.group(1))
        inches=int(match_ftich.group(2))
        total_inches=(feet*12)+inches
        cm=total_inches*2.54
        return int(cm)
    match_cm=re.match(r'(\d+)cm',height)
    if match_cm:
        return int(match_cm.group(1))
    return None
```

```python
df['Height']=df['Height'].apply(height_converter)
```

```python
def weight_converter(weight):
    match_lbs=re.match(r'(\d+)lbs', weight)
    if match_lbs:
        lbs=int(match_lbs.group(1))
        kg=lbs*0.453592
        return int(kg)
    match_kg=re.match(r'(\d+)kg',weight)
    if match_kg:
        return int(match_kg.group(1))
    return None
```

```python
df['Weight']=df['Weight'].apply(weight_converter)
df['Weight'].head(10)
```
